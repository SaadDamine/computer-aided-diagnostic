```python
# importing necessary python packages
import numpy as np
from sklearn import neighbors
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
df = pd.read_csv('breast-cancer-wisconsin.data',header=None)
df.replace('?',-99999,inplace=True)
df
```

```python
#assign column names to the dataset as follows
columns=['id','Clump Thickness','Uniformity of Cell Size',
            'Uniformity of Cell Shape','Marginal Adhesion','Single Epithelial Cell Size','Bare Nuclei','Bland Chromatin',
           'Normal Nucleoli','Mitoses','Class']
```

```python
df.columns=columns
df
```

```python
x=np.array(df.drop(['Class'],axis=1))
y=np.array(df['Class'])
```

```python
#ivide the data into train and test split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#rain the model with the help of KNeighborsClassifier 
clf= neighbors.KNeighborsClassifier()
clf.fit(x_train,y_train)
accurancy=clf.score(x_test,y_test)
#print("accuracy",accurancy)

#make prediction
y_pred = clf.predict(x_test)

# print the results as follows
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
result = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(result)
result1 = classification_report(y_test, y_pred)
print("Classification Report:",)
print (result1)
result2 = accuracy_score(y_test,y_pred)
print("Accuracy:",result2)
```

```python
exemple_mesure=np.array([[114676,2,1,2,1,2,2,3,1,1]])
prediction=clf.predict(exemple_mesure)
print("new Prediction",prediction) #(2 for benign, 4 for malignant)
```
